3. Tek Değişkenli Analiz (Univariate Analysis)


```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sleep Disorder', data=df, palette='viridis')
plt.title('Uyku Bozukluğu Dağılımı')
plt.xlabel('Uyku Bozukluğu')
plt.ylabel('Birey Sayısı')
plt.show()

print("\nUyku Bozukluğu Frekansları:")
print(df['Sleep Disorder'].value_counts(normalize=True))
```

Uyku bozukluğu dağılım grafiğini çizdik uyku bozukluğu olmayanlar %58 Uyku Apnesi olanlar %20 Insomnia ise %20 civarlarında görülmektedir

Sayısal Değişkenlerin Dağılımı

```python
# Uyku Süresi
plt.subplot(1, 3, 1)
sns.histplot(df['Sleep Duration'], kde=True, bins=10, color='skyblue')
plt.title('Uyku Süresi  (Saat)')
plt.xlabel('Uyku Süresi')

# Uyku kalitesi
plt.subplot(1, 3, 2)
sns.countplot(x='Quality of Sleep', data=df, palette='light:green')
plt.title('Uyku Kalitesi  (1-10)')
plt.xlabel('Uyku Kalitesi')

# Stres Seviyesi
plt.subplot(1, 3, 3)
sns.countplot(x='Stress Level', data=df, palette='light:salmon')
plt.title('Stres Düzeyi (1-10)')
plt.xlabel('Stres Düzeyi')

plt.tight_layout()
plt.show()
```

Uyku Süresi: 6.5 saat civarı ve 7.5−8 saat civarında yoğunlaşmalar var.

Uyku Kalitesi: 8 en yaygın kalite puanı, ardından 6 ve 7 geliyor. 4 ve 5 puanları düşük uyku kalitesine işaret ediyor.

Stres Düzeyi: 3 ve 8 seviyeleri en yaygın stres düzeyleri.

Kategorik Değişkenlerin Dağılımı

```python
plt.figure(figsize=(18, 5))

# Cinsiyet
plt.subplot(1, 3, 1)
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Cinsiyet Dağılımı')
plt.xlabel('Cinsiyet')

# Vücut kitle indeksi
plt.subplot(1, 3, 2)
sns.countplot(x='BMI Category', data=df, palette='Set2')
plt.title('BKİ Kategorisi Dağılımı')
plt.xlabel('BKİ Kategorisi')
plt.xticks(rotation=15)

# Meslek (İlk 10 meslek)
plt.subplot(1, 3, 3)
sns.countplot(y='Occupation', data=df, order=df['Occupation'].value_counts().index, palette='tab10')
plt.title('Meslek Dağılımı')
plt.ylabel('Meslek')

plt.tight_layout()
plt.show()
```

Cinsiyet nerdeyse eşit dağılmış.
Bki yani vücut kitle indeksinin çoğunluğu normalken bunu fazla kilolu grubu takip etmiş.Mesleklerde ise en fazla hemşire varken bunu doktorlar takip etmiş

4. İKİ VEYA DAHA FAZLA DEĞİŞKEN ANALİZİ

```python
comparison_cols = ['Sleep Duration', 'Quality of Sleep', 'Stress Level', 'Daily Steps']

plt.figure(figsize=(15, 10))
for i, col in enumerate(comparison_cols):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x='Sleep Disorder', y=col, data=df, palette='coolwarm')
    plt.title(f'{col} ve Uyku Bozukluğu İlişkisi')
    plt.xlabel('Uyku Bozukluğu')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

Insomnia uyku bozukluğunun en düşük uyku süresiyle ilişkili olduğunu görüyoruz.İnsomnia ve Uyku Apnesi olanların uyku kalitesi puanları, No Disorder grubuna göre çok daha düşüktür. Özellikle İnsomnia grubunda ortalama 4−5 civarında
Insomnia olanların stres seviyelerinin en yüksek olduğu görülüyor
No Disorder grubunun ortalama adım sayısı, İnsomnia ve Uyku Apnesi gruplarına göre daha yüksektir. Bu, fiziksel aktivitenin olumlu etkisini düşündürebilir.

```python
plt.figure(figsize=(15, 6))

# BMI Category ve Sleep Disorder
plt.subplot(1, 2, 1)
sns.countplot(x='BMI Category', hue='Sleep Disorder', data=df, palette='Set1')
plt.title('BKİ Kategorisine Göre Uyku Bozukluğu Dağılımı')
plt.xlabel('BKİ Kategorisi')
plt.ylabel('Birey Sayısı')
plt.xticks(rotation=15)
plt.legend(title='Uyku Bozukluğu')

# Occupation ve Sleep Disorder (İlk 8 meslek)
top_occupations = df['Occupation'].value_counts().head(8).index
df_top_occ = df[df['Occupation'].isin(top_occupations)]

plt.subplot(1, 2, 2)
sns.countplot(y='Occupation', hue='Sleep Disorder', data=df_top_occ, order=top_occupations, palette='Set1')
plt.title('Mesleğe Göre Uyku Bozukluğu Dağılımı (Önde Gelenler)')
plt.ylabel('Meslek')
plt.xlabel('Birey Sayısı')
plt.legend(title='Uyku Bozukluğu')

plt.tight_layout()
plt.show()
```

BKİ Kategorisi (BMI Category):

    Uyku Apnesi vakalarının büyük çoğunluğu Overweight (Fazla Kilolu) ve Obese (Obez) kategorilerinde yoğunlaşmıştır. Normal BKİ kategorisindeki bireylerde Uyku Apnesi görülme olasılığı düşüktür.

    İnsomnia vakaları Normal ve Overweight kategorilerinde daha yaygındır.

    No Disorder vakalarının çoğunluğu Normal BKİ kategorisindedir.

Meslek (Occupation):

    Sales Representative (Satış Temsilcisi) mesleği neredeyse tamamen İnsomnia veya Uyku Apnesi vakalarıyla ilişkilidir. No Disorder bu grupta çok azdır.

    Teacher (Öğretmen) mesleğinde İnsomnia ve Uyku Apnesi vakaları, No Disorder vakalarıyla benzer orandadır.

    Nurse (Hemşire) ve Doctor (Doktor) mesleklerinde de uyku bozuklukları belirgin bir şekilde görülmektedir.

    Engineer (Mühendis) ve Accountant (Muhasebeci) gibi mesleklerde No Disorder
    vakaları daha baskındır, bu da bu meslek gruplarının genel olarak daha iyi uyku sağlığına sahip olabileceğini düşündürür.

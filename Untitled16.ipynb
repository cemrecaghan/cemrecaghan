**bu değişken için z score yöntemi ile aykırı değer tespiti yapınız.**

```python
upper_w = df_clean['price'].quantile(0.95)
upper_w
```

**elde edilen yeni değerlere göre price değişkenin tanımlayıcı istatistiklerini elde

elde edilen yeni değerlere göre price değişkenin histogramını elde ediniz ve
sonucu yorumlayınız.**


```python
df["price_w"] = df["price"].clip(upper=upper_w)
df["price_w"].describe()
```

```python
sns.histplot(df["price_w"], bins=50)
plt.title("Winsorize Sonrası Price Histogramı")
plt.show
```

```python
### 2.2. NİCEL (SAYISAL) DEĞİŞKENLERİN ANALİZİ


#bu değişkenli ilgili sayısal özet bilgileri elde ediniz (ortalama, median, varyans ..gibi) ve yorumlayız

import pandas as pd
df = pd.read_csv("clean_books.csv",sep=";")
df
df.describe()
```

```python
#yıllara göre basılan kitapları listeleyiniz
df['year'].value_counts().sort_index()
```

```python
#- Kitap yayınları yıllar içinde nasıl bir trend gösteriyor? yorumlayınız (çizgi(line) plot çizerek yorumlayabilirisniz)
import matplotlib.pyplot as plt
import seaborn as sns

kitap_yil = df['year'].value_counts().sort_index()
sns.lineplot(x=kitap_yil.index,y=kitap_yil.values)
plt.xlabel=("kitap_yil")
plt.ylabel=("basılan kitap sayısı")
plt.title=("Yıllara göre basılan kitap sayısı")
plt.show()
```

```python
# rating değeri maximum kaçtır ve bu değeri alan kitapları listeleyiniz ve bunu yeni bir df olarak kaydedizniz.
max_rating=df['rating'].max()
max_rating
```

```python
#en yüksek değeri alan kitaplar
enyüksek_degerli_kitaplar=df[df['rating']==max_rating]
enyüksek_degerli_kitaplar
```

```python
enyüksek_degerli_kitaplar_df = enyüksek_degerli_kitaplar.copy()
kurgu_kitaplari = enyüksek_degerli_kitaplar_df[enyüksek_degerli_kitaplar_df['genre'] == 'Fiction']
cocuk_kitaplari = enyüksek_degerli_kitaplar_df[enyüksek_degerli_kitaplar_df['genre'] == 'Children']
kurgudisi_kitaplari = enyüksek_degerli_kitaplar_df[enyüksek_degerli_kitaplar_df['genre'] == 'Non-Fiction']

print("Kurgu Kitapları:\n", kurgu_kitaplari)
print("Çocuk Kitapları:\n", cocuk_kitaplari)
print("Kurgudışı Kitapları:\n", kurgudisi_kitaplari)
```

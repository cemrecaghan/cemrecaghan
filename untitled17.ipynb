**Kayıp veri içeren kategorik değişkenler için uygun yöntem kullanarak imputasyon
işlemi yapınız**


```python
from scipy.stats import skew

imputed_df = df.copy()

for col in numeric_cols:
    non_null = df[col].dropna()

    # Eğer tüm değerler NaN ise atla
    if len(non_null) == 0:
        print(f"{col}: Tüm değerler eksik, imputasyon yapılamadı.")
        continue

    # Eğer tek değer varsa skew hesaplanamaz → o değeri doldur
    if len(non_null) == 1:
        value = non_null.iloc[0]
        imputed_df[col] = df[col].fillna(value)
        print(f"{col}: Tek değer vardı, aynı değerle dolduruldu ({value}).")
        continue

    s = skew(non_null)

    if s < 1:
        imputed_value = df[col].mean()
        method = "Ortalama (Mean)"
    else:
        imputed_value = df[col].median()
        method = "Medyan (Median)"

    imputed_df[col] = df[col].fillna(imputed_value)
    print(f"{col} → {method} ile dolduruldu | Çarpıklık: {s:.3f}")
```

```python
stats_before = df[numeric_cols].describe().T
stats_after = imputed_df[numeric_cols].describe().T

print("\n----- İmputasyon Öncesi İstatistikler -----")
display(stats_before)

print("\n----- İmputasyon Sonrası İstatistikler -----")
display(stats_after)
```

```python
Kayıp veri içeren kategorik değişkenler için uygun yöntem
kullanarak imputasyon işlemi yapınız
```

```python
cat_cols = df.select_dtypes(include=["object","category"]).columns

print("\n----- Kategorik Değişken İmputasyonu (Mod) -----")
for col in cat_cols:
    if df[col].isna().sum() == 0:
        continue
    mode_val = df[col].mode()[0]
    imputed_df[col] = df[col].fillna(mode_val)
    print(f"{col}: Mod değer ({mode_val}) ile dolduruldu.")
```

Kategorik değişkenlerde imputasyon öncesi–sonrası frekans karşılaştırması

```python
for col in cat_cols:
    print(f"\n----- {col} Frekans (Önce) -----")
    display(df[col].value_counts(dropna=False))

    print(f"----- {col} Frekans (Sonra) -----")
    display(imputed_df[col].value_counts(dropna=False))
```


Mod ile doldurma, kategorik değişkenler için temel ve yaygın bir imputasyon yöntemidir.

**Ekstra ki-kareyi de ekledim**

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Veri setini yükleme ve temizleme (Önceki adımdan)
df = pd.read_csv('saglik_sigortasi_veri_seti.csv')
df_chi = df.drop('gunluk_sigara', axis=1).copy() # %100 kayıp olanı çıkardık

# Sadece tam (non-missing) gözlemleri kullanarak ilişkiyi test etme
# Her iki sütunda da NaN olmayan satırları seçiyoruz.
contingency_table = pd.crosstab(df_chi['meslek'], df_chi['egitim_durumu'], dropna=True)

# Ki-Kare testi uygulama
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("### Ki-Kare Bağımsızlık Testi Sonuçları: meslek vs. egitim_durumu")
print(f"Ki-Kare İstatistik Değeri (chi2): {chi2:.2f}")
print(f"p-Değeri (p-value): {p:.5f}")
print(f"Serbestlik Derecesi (dof): {dof}")
```



  p-Değeri: Eğer p-değeri, kabul edilen anlamlılık seviyesi olan 0.05'ten küçükse (p<0.05), H0​ hipotezi reddederiz.

  Sonuç: Genellikle meslek ve egitim_durumu gibi sosyoekonomik değişkenler arasında çok güçlü bir ilişki beklenir (örneğin p<0.0001). Bu durumda, H0​ reddedilir ve bu iki değişkenin birbiriyle bağımlı olduğu sonucuna varılır.
